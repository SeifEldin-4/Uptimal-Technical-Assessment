```python
import pandas as pd
import requests
```

```python
df_holidays = []

for year in range(2012, 2025):  
    url = f"https://date.nager.at/api/v3/publicholidays/{year}/US"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        df = pd.DataFrame(data)
        df_holidays.append(df)
    else:
        print(f"Error: {response.status_code}")

df_holidays = pd.concat(df_holidays, ignore_index=True)
df_holidays_NY = df_holidays[df_holidays['counties'].apply(lambda x: any('NY' in c for c in x) if isinstance(x, list) else False) | df_holidays['global'] == True]
df_holidays_NY = df_holidays_NY[['date','name']]
```

```python
url = "https://data.cityofnewyork.us/api/views/h9gi-nx95/rows.csv?accessType=DOWNLOAD"
df_collisions = pd.read_csv(url)
```

```python
df_collisions['CRASH DATE'] = pd.to_datetime(df_collisions['CRASH DATE'])
df_holidays_NY['date'] = pd.to_datetime(df_holidays_NY['date'])
```

```python
df_collisions = df_collisions.merge(df_holidays_NY,how='left',left_on='CRASH DATE',right_on='date')
df_collisions.drop(columns='date',inplace=True)
#df_collisions[df_collisions['name'].notna()]
```

```python
df_collisions['ZIP CODE'] = df_collisions['ZIP CODE'].astype(str)
df_collisions.to_parquet("collisions_data.parquet", index=False)
```
